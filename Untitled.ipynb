```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras import Sequential
from keras.layers import Dense,Dropout
```

```python
data=pd.read_csv("./Train/Train_Data.csv")

data.columns
```

```python
data.describe().T
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), cmap="RdBu")
plt.title("Correlations Between Variables", size=15)
plt.show()
```

```python
important_num_cols = list(data.corr()["SalePrice"][(data.corr()["SalePrice"]>0.50) | (data.corr()["SalePrice"]<-0.50)].index)
```

```python
cat_cols = ["MSZoning", "Utilities","BldgType","Heating","KitchenQual","SaleCondition","LandSlope"]
important_cols = important_num_cols + cat_cols

data = data[important_cols]
```

```python
print("Missing Values by Column")
print("-"*30)
print(data.isna().sum())
print("-"*30)
print("TOTAL MISSING VALUES:",data.isna().sum().sum())
```

```python
sns.pairplot(data[important_num_cols])
```

```python
X = data.drop("SalePrice", axis=1)
Y = data["SalePrice"]
```

```python
X.head()
```

```python
x = pd.get_dummies(X, columns=cat_cols)
x.head()
```

```python
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler

important_num_cols.remove("SalePrice")

scaler = StandardScaler()
X[important_num_cols] = scaler.fit_transform(X[important_num_cols])
```

```python
X=X.values

Y=Y.values
```

```python
X = np.asarray(X).astype('float32')
Y = np.asarray(Y).astype('float32')
```

```python
Y
```

```python
X
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
model=Sequential()
```

```python
model.add(Dense(78,activation="relu",input_shape=(39,)))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Dense(156,activation="relu"))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Dense(13,activation="relu"))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Dense(1,activation="linear"))
```

```python
opi=SGD(learning_rate=0.001,clipnorm=1)
```

```python
model.compile(loss='mean_squared_error',optimizer=opi,metrics=['accuracy', 'mse'])
```

```python
model.summary()
```

```python
hist=model.fit(X,Y,batch_size=10,epochs=1000,validation_split=0.2)
```

```python
model.predict(X).reshape(-1)
```

```python
Y
```

```python
plt.plot(hist.epoch[50:],hist.history["loss"][50:])
```
